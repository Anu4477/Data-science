```python
import pandas as pd
#import panda
df= pd.Series([10,20,30])
print(df)
print(df[0])
print(df[1])
print(df[2])

print(df.index)
print(df.values)
```

```python
import pandas as pd

data = {
    "name": ["hello", "kunal", "praveen", "anjali", "abhishek"],
    "age": [20, 21, 22, 23, 24],
    "salary": [
        {"In hand": "15000", "ctc": "1.2lpa"},
        {"In hand": "20000", "ctc": "2lpa"},
        {"In hand": "25000", "ctc": "3lpa"},
        {"In hand": "30000", "ctc": "4lpa"},
        {"In hand": "35000", "ctc": "5lpa"}
    ]
}

df = pd.DataFrame(data)
print(df)
```

```python
#dict
import pandas as pd
dict = {"name":["dheeraj","kunal","praveen","anjali","abhishek singh"],
        "age":[20,21,22,23,30],
        "salary":[15000,20000,25000,30000,35000]
        }
df = pd.DataFrame(dict)
print(df)
```
