```python
#PRACTICE QUESTION
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data)

print(df)

print("--------------------------------------------")

# Head
print(df.head(3))

print("--------------------------------------------")

# Tail
print(df.tail(3))

print("--------------------------------------------")

# Info
print(df.info())


print("--------------------------------------------")

# Rename Column
df.rename(columns={"Salary": "Monthly_Salary"}, inplace=True)
print(df)

print("--------------------------------------------")

# Shape
print(df.shape)

print("--------------------------------------------")

# Describe
print(df.describe())

```
