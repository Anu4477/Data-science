OPERATIONS ON ROW

```python
import pandas as pd
d={
    "name":["rahul","anushka","sini"],
    "salary":[100,200,300]
}
df=pd.DataFrame(data=d)
print(df.loc[2,"name"])
print(df.iloc[2,0])
```

```python
#get single row using loc
import pandas as pd
d={
    "name":["rahul","anushka","sini"],
    "salary":[100,200,300]
}
df=pd.DataFrame(data=d)
#get single row using iloc
print(df.iloc[0])
#get single row using loc
print(df.loc[0])
#get multi row using loc
print(df.loc[0:1])
#get multi row using iloc
print(df.iloc[[0,1]])
#sub data get using loc
df1=df.iloc[0:2,[0]]
df1
#sub data get using loc
df2=df.loc[0:2,["name"]]
df2
```

```python
#PRACTICE QUESTION
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data)
#add column
df["incentives"]=df["Salary"]*0.1
#delete column
df.drop(["Salary"],axis=1,inplace=True)
#get single row using iloc
print(df.iloc[0])
#get single row using loc
print(df.loc[0])
#get multi row using loc
print(df.loc[0:1])
#get multi row using iloc
print(df.iloc[[0,1]])
```

```python
import pandas as pd
url="https://raw.githubusercontent.com/Anu4477/Data-science/main/studentdata.json"
df= pd.read_json(url)
#add columns
df["hindi"]=df["maths"]+30
#remove column
df.drop(["maths"],axis=1,inplace=True)
#get single row using iloc
print(df.iloc[0])
#get single row using loc
print(df.loc[0])
#get multi row using loc
print(df.loc[0:1])
#get multi row using iloc
print(df.iloc[[0,1]])
#2 specific value by using jump
print(df.iloc[[0,1],[0,1,2]])
#
df
```
