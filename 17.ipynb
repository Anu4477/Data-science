```python
import pandas as pd
url="https://raw.githubusercontent.com/Anu4477/Data-science/main/file2.json"
df = pd.read_json(url)
df['doj']=['2025-01-01','2025-01-02','2025-01-03','2025-01-04','2025-01-05']
df

#convert string to date
df['doj']=pd.to_datetime(df['doj'])
df['doj'].dtypes

#date operation
df['doj'].dt.year
df['doj'].dt.month
df['doj'].dt.day
df['doj'].dt.day_name()

#20 days
df['doj']+pd.to_timedelta(20,unit='D')
```
