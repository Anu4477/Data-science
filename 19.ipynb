**Aggrregation and group by**

```python
#aggregation
import pandas as pd
url ="https://raw.githubusercontent.com/Anu4477/Data-science/main/file2.json"
df=pd.read_json(url)
print(df)
df['marks'].mean()
df['marks'].sum()
df['marks'].min()
df['marks'].max()

df['marks'].agg(["sum","mean","min","max"])
```
