MULTIPLE LINE CHART

```python
#Multi lines chart
import matplotlib.pyplot as plt
x=[2010,2015,2020,2025]
y1=[100,200,250,300]
y2=[150,250,300,250]
plt.plot(x,y1,label='jeans',color='pink')
plt.plot(x,y2,label='tops',color='green')
plt.legend()
plt.xlabel('year')
plt.ylabel('sales')
plt.title('sales report')
plt.show()
```

```python
#multiple line chart
import matplotlib.pyplot as plt
x=[2005,2010,2015,2020,2025]
y1=[10,20,32,30,45]
y2=[8,12,35,33,25]
y3=[15,10,30,44,50]
plt.plot(x,y1,label='shirt',marker='*',color='yellow',markersize =4,linestyle='--',linewidth=2)
plt.plot(x,y2,label='jeans',marker='*',color='grey',markersize =4,linestyle='--',linewidth=2)
plt.plot(x,y3,label='top',marker='*',color='pink',markersize =4,linestyle='--',linewidth=2)
plt.xlabel('year')
plt.ylabel('sales')
plt.title('sales report')
plt.legend()
plt.show()
```

BAR CHART

```python
#bar chart
import matplotlib.pyplot as plt
x=[2005,2010,2015,2020,2025]
y=[10,20,32,30,45]
plt.bar(x,y,color='pink')
plt.xlabel('year')
plt.ylabel('sales')
plt.title('sales report')
plt.show()

```

```python
#multiple line chart
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y1=[10,20,30,40]
y2=[20,20,25,30]

#calculation for width
w=0.40
plt.bar(x-w/2,y1,label="girls",width=w)
plt.bar(x+w/2,y2,label="boys",width=w)
plt.xlabel('group')
plt.ylabel('no. of students')
plt.title('student report')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

groups = ['Group A','Group B','Group C','Group D']
y1 = [10,20,30,40]
y2 = [20,20,25,30]

x = np.arange(len(groups))  # [0,1,2,3]

w = 0.40

plt.bar(x-w/2, y1, width=w, label="Girls")
plt.bar(x+w/2, y2, width=w, label="Boys")

plt.xticks(x, groups)

plt.xlabel('Group')
plt.ylabel('No. of Students')
plt.title('Student Report')
plt.legend()

plt.show()
```
