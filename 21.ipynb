```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2001, 2002, 2003, 2004, 2005])
y1 = [10, 20, 30, 40, 50]
y2 = [12, 22, 32, 42, 48]
y3 = [15, 21, 33, 46, 56]
y4=[16,25,38,44,50]
w = 0.2
plt.bar(x - w, y1, label="car", width=w)
plt.bar(x,y2,  label="bike", width=w)
plt.bar(x + w, y3,label="EV", width=w)
plt.bar(x + w*2, y4,label="truck", width=w)
plt.xlabel("year")
plt.ylabel("sales")
plt.title("shop report")
plt.legend()
plt.show()
```

```python

#pie chart
import matplotlib.pyplot as plt
fruits=['apple','banana','orange','watermelon']
count=[20,30,40,50]
colors = ['red','yellow','orange','green']
plt.pie(count ,labels=fruits,colors=colors)
#plt.plot(count,fruits)
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
fruits=['apple','banana','orange','watermelon']
count=[20,30,40,50]
colors = ['red','yellow','orange','green']
plt.pie(count ,labels=fruits,colors=colors,startangle=180,autopct="%1.1f%%")
plt.show()
```

```python
#subplots
#first chart
x=[1,2,3,4]
y=[10,20,30,40]
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.subplot(1,2,1)
plt.show()
#second chart
x=[5,6,7,8]
y=[50,60,70,80]
plt.plot(x,y)
plt.xlabel("")
plt.ylabel("")
plt.subplot(1,2,2)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Chart 1
plt.subplot(2,2,1)
x = [1,2,3,4,5]
y = [10,20,30,40,55]
plt.plot(x,y)
plt.title("Line Chart")

# Chart 2
plt.subplot(2,2,2)
fruits = ['apple','banana','orange','watermelon']
sales = [40,30,15,70]
plt.pie(sales, labels=fruits, autopct="%1.1f%%")
plt.title("Pie Chart")

# Chart 3
plt.subplot(2,2,3)
x = [10,20,20,30,30,30,40,40,50]
plt.hist(x,bins=4)
plt.title("Histogram")

# Chart 4
plt.subplot(2,2,4)
x = ['A','B','C','D']
y = [15,25,35,45]
plt.bar(x,y)
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```
