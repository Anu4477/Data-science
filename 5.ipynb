````python
#nested dictionary
d={"address":{"address1":{"city":"kukas","district":"jaipur"},
              "address2":{"city":"gopalpura","district":"arya mains"}}}
print(d)
print(d["address"]["address1"]["city"])
print(d["address"]["address1"]["district"])
print(d["address"]["address2"]["city"])
print(d["address"]["address2"]["district"])

print(d["address"]["address1"])
print(d["address"]["address2"])

for i in d.values():
    for j in i.values():
        for k in j.values():
            print(k)

````

Nested list
1.slicing

````python
#example
l=[10,20,30,["hello","hello1","hello2",[True,False]]]

l1 = l[-3:-1]
print(l1)

print(l[3][1:3])



print(l[3][3][0])
print(l[3][3][1])
print(l[3][2])
print(l[3][1])
print(l[3][0])
print(l[2])
print(l[1])
print(l[0])



for i in range(len(l)):
    if type(l[i]) == list:
        for j in range(len(l[i])):
            if type(l[i][j]) == list:
                for k in range(len(l[i][j])):
                    print(l[i][j][k])
            else:
                print(l[i][j])
    else:
        print(l[i])
````
