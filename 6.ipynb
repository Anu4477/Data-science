Importing and using modules

```python
#import module
import math
print(math.factorial(5))
print(math.pow(5,2))
print(math.pi)
print(math.floor(10.7234))
print(math.ceil(10.7234))

print("--------------------")

#example
import math as anushka
print(anushka.factorial(5))
print(anushka.pow(5,2))
print(anushka.pi)
print(anushka.floor(10.7234))
print(anushka.ceil(10.7234))

print("--------------------")


#example2
from math import factorial,pi
print(factorial(5))
print(pi)

print("--------------------")

#example3
from math import *
print(factorial(5))
print(pi)
print(floor(10.7234))
print(ceil(10.7234))
```
