```python
#example
class anushka:
  def __init__(self,name):
    self.name = name
  def show(self):
    print(self.name)
a = anushka("hello")
a.show()

print("---------------------")

#example2
class anushka:
  def __init__(self):
    print("calling constructor")

  def show(self):
    print("show the name:")

r = anushka()
r.show()

print("---------------------")

#example3
class anushka:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def getage(self):
    print("my age is :",self .age)

  def getname(self):
    print("my name is :",self.name)

a = anushka("anushka",20)
a.getname()
a.getage()

print("---------------------")

#example4
class student:
  def __init__(self,*args):
    print(type(args))
    print(args)
    self.name=args[0]

  def getstu(self):
    print("the student is:",self.name)

s = student("anushka",20,"000000000","arya@gmail.com")
t= s.getstu()
print(t)

print("---------------------")
#example4
class student:
    def __init__(self, **args):
        print(type(args))
        print(args)

        self.name = args["name"]
        self.age = args["age"]
        self.phone = args["phone"]
        self.email = args["email"]

    def getstu(self):
        print("The student is:", self.name)

s = student(
    name="anushka",
    age=20,
    phone="000000000",
    email="arya@gmail.com"
)

s.getstu()

print("---------------------")

#example5
class student:
  def __init__(self,*args):
    print(type(args))
    print(args)
    self.name=args[0]
    self.age=args[1]
    self.phone=args[2]
    self.email=args[3]

  def getstu(self):
    print("the student is:",self.name)

s = student("anushka",20,"000000000","arya@gmail.com")
s.getstu()


print("-------------------")

#example6
class student:
    def __init__(self, *args):
        self.data = args

    def users(self):
        for user in self.data[0]:
            print(user)

    def details(self):
        for key, value in self.data[1].items():
            print(key, ":", value)


s = student(
    ["dheeraj", "kunal", "nikita", "pinky"],
    {"address": "kukas", "college": "arya", "loc": "jaipur"}
)

s.users()
print()
s.details()

```
